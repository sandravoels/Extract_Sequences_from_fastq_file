```python
grouped = df.groupby('peptide')['sample']
```

```python
list_peptide=[]
list_count=[]
list_members=[]

for name, group in grouped:
    list_peptide.append(name)
    list_count.append(group.count())
    list_members.append(group)
    
print(len(list_peptide))
print(len(list_count))
print(len(list_members))

```

```python
df = pd.DataFrame(list(zip(list_peptide, list_count,list_members)),
               columns =['peptide', 'Count', 'Members'])
df
```
