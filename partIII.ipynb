```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r"C:\Users\JonathanG03\Dropbox\MotifAi_Exercises\26A_lakshmi\exp_26A.xlsx","short_data")
```

```python
#columns "positive": if average mAB bigger than average naive
def fxy(x, y):
    if x>y:
        return "Big"
    else:
        return "Small"

 
df['Positive']=df.apply(lambda x: fxy(x['average_mAB'], x['average_naive']), axis=1)
```

```python
#drop line that "positive"==small
df.drop(df.index[df['Positive'] == 'Small'], inplace = True)
```

```python
#replace 0 to NaN
df=df.replace(0,np.nan)
```

```python
#drop line with NaN in specific columns
df.dropna(subset = ["m336","m337","m338",'C12','F11','JC57-11','CDC2-C2'], inplace=True)
```
