Note the leftmost column in all of the above - those are the labels of the of the rows. Labels are not the same as positional indexes! They can be numbers, and by default they are in order, but this can change as the DataFrame changes (examples to follow). Make sure to understand what you're working with!

In Pandas jargon, a 'label' and 'index' are interchangeable, and are like keys in a dictionary (though they don't have to be unique). Python's list indices are called 'positional indices' in Pandas, and we will adopt this naming throughout the notebook.

Accessing data in Pandas can be a bit complicated. If unsure, consult the special indexing docs: http://pandas.pydata.org/pandas-docs/stable/indexing.html

The above warning is generated because `subunits` can be either a "view" or a complete copy of the column. If it is a view, changing it will change the original DataFrame it came from (`df`), as intended. But if it is a copy, then only the copy will change, and not `df`! Pandas does not guarantee whether a view or a copy will be returned when using the `[]` syntax, and thus assignment should be done with `.loc[]` or `.iloc[]` (or `.at[]` and `.iat[]`, but we don't discuss those here).

```python
df.loc[55, 'Subunits'] = old_value
display(df.loc[53:57])
# Notice how no warning is generated now, as we are modifying 'df' directly - there's no ambiguity.
```

```python
subunits_copy[55] = 6
display(df.loc[53:57])
# There's no warning here, either - that's because subunits_copy is an explicit copy of subunits,
# so Pandas knows that we know that df won't change - and thus there's no hidden bug here
# to warn us about.
```

# Iterating over records 

```python
charge_sum = 0

for i in range(len(df)):
    record = df.iloc[i]
    charge_sum += record['Net Surface Charge']
    
print(charge_sum)
print(charge_sum == df['Net Surface Charge'].sum())
```

```python
charge_sum = 0

for index, record in df.iterrows():
    charge_sum += record['Net Surface Charge']
    
print(charge_sum)
print(charge_sum == df['Net Surface Charge'].sum())
```

```python
def option1():
    
    charge_sum = 0

    for i in range(len(df)):
        record = df.iloc[i]
        charge_sum += record['Net Surface Charge']
        
    return charge_sum
    
def option2():
    
    charge_sum = 0

    for index, record in df.iterrows():
        charge_sum += record['Net Surface Charge']
        
    return charge_sum

%timeit option1()
%timeit option2()
%timeit df['Net Surface Charge'].sum()
```

```python
series = df['Net Surface Charge']

charge_sum = 0

for charge in series:
    charge_sum += charge
    
print(charge_sum)
print(charge_sum == series.sum())
```

```python
charge_sum = 0

for index, charge in series.iteritems():
    charge_sum += charge
    
print(charge_sum)
print(charge_sum == series.sum())
```

# Applying functions on records

```python
def parse_resolution(raw_resolution):
    try:
        return float(raw_resolution)
    except ValueError:
        return np.nan
        
print(df['Resolution'].apply(parse_resolution).sample(15)) # .apply() returns a new Series
```

```python
df['Resolution'] = df['Resolution'].apply(parse_resolution)
display(df)
```

```python
def get_taxonomy(record):
    try:
        return '%s/%s/%s' % (record['Genome'].split()[0], record['Family'], record['Genus'])
    except:
        return np.nan
    
df['Taxonomy'] = df.apply(get_taxonomy, axis = 1)
display(df)
```

```python
# applymap works element-wise
display(df[['Inner Radius', 'Outer Radius', 'Average Radius']].applymap(lambda x: 'Big' if x >= 200 else 'Small'))
```

```python
# pd.isnull is different (and usually better) than np.isnan

for value in [np.nan, 0, None, '']:
    
    for function in [pd.isnull, np.isnan]:
        try:
            print('%s(%s) = %s' % (function.__name__, repr(value), str(function(value))))
        except Exception as e:
            print('%s(%s) raises an error (%s)' % (function.__name__, repr(value), str(type(e))))
    
    print('*' * 20)
```
