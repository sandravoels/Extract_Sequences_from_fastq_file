```python
df = df[df['both'].notna()]
df
```

```python
df_50_five= pd.read_csv('Top_50_features_full.csv')
#display(df_50_short)

list_of_discriminatory_motifs=df_50_five.columns.values.tolist()[1:]

list_of_discriminatory_motifs
```

```python
def find_disc_motifs(row):
    for motif in list_of_discriminatory_motifs:
        sum_full=row.str.count(motif).sum()
        if sum_full >=1:
            return motif

df['Motif'] = df.apply(find_disc_motifs, axis=1)
df
# print (df['Motif 1'])

# if '' in df['Motif 1']:
#     print ('yes')
```
